```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.arange(1,5,0.1)
y = x**3
plt.plot(x,y, color="red", marker="o")
```

```python
plt.plot(x,y, color="red", marker="p")
```

```python
plt.plot(x,y, color="red", linewidth = 0.5)
```

```python
plt.plot(x,y, color="red", linewidth = 5)
```

if you want label in x-axis and y-axis

```python
x = np.arange(1,5,0.1)
y = x**3
y2 = x**2
plt.plot(x, y, color="red", linewidth = 5)
plt.plot(x, y2, color="blue", linewidth = 5)

plt.xlabel("x")
plt.ylabel("x^3")
plt.title("Matplotlib demo")
```

```python
#color, marker, linewidth, labels for axis, datasets and the title
x = np.arange(1,5,0.1)
y = x**3
y2 = x**2

plt.plot(x, y, color="red", linewidth = 5, label = "x^3")
plt.plot(x, y2, color="blue", linewidth = 5, label = "x^2")

plt.xlabel("x")
plt.ylabel("y")

plt.legend()

plt.title("Matplotlib demo")
```
