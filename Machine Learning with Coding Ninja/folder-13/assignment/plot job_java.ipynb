```python
import csv
import matplotlib.pyplot as plt
```

```python
import collections
```

```python
with open("amazonjobsdataset.csv") as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace=True)
    dct = {}
    for row in file_data:
        key1 = row['BASIC QUALIFICATIONS']
        date = row['Posting_date'].split()
        key = date[2]
        string = "Java"
        ans = key1.find(string)
        if ans != -1:
            if key in dct:
                dct[key] += 1
            else:
                dct[key] = 1
    ord_dct = collections.OrderedDict(sorted(dct.items()))                                  
    plt.scatter(list(ord_dct.keys()), list(ord_dct.values()))
    plt.xlabel("Year")
    plt.ylabel("Jobs")
    plt.title('year vs No. of jobs opening related to Java')
    plt.xticks(rotation = 40)
    plt.show()
    
for i in ord_dct.keys():
    print(i,ord_dct[i])
```

```python
import csv
import matplotlib.pyplot as plt
import collections
with open('amazonjobsdataset.csv', encoding ='UTF-8') as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace=True)

    dct = {}
    for row in file_data:
        qlfn = row['BASIC QUALIFICATIONS']
        if 'Java' in qlfn or 'java' in qlfn :
            key = row['Posting_date'].split()[2]
            if key in dct:
                dct[key] += 1
            else:
                dct[key] = 1
            
    ord_dct = collections.OrderedDict(sorted(dct.items()))                                  
    plt.plot(list(ord_dct.keys()),list(ord_dct.values()))
    
    plt.xlabel("Year")
    plt.ylabel("Jobs")
    plt.title('year vs No. of jobs opening related to Java')
    plt.xticks(rotation = 40)
    plt.show()
    
    for i in ord_dct.keys():
        print(i,ord_dct[i])
```
