```python
import csv
import matplotlib.pyplot as plt
import collections
```

```python
with open("amazonjobsdataset.csv") as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace=True)
    dct = {}
    for i in file_data:
        city = i['location']
        city1 = i['location'].split(',')[0]
        if city1!='IN':
            continue
            
        ans = ""
        ans1 = ans+city
        ans2 = ans1.split(',')
        key1 = ans2[0]
        key2 = ans2[2]
        if key2 in dct:
            dct[key2] += 1
        else:
            dct[key2] = 1
#         print(key2)       
    plt.pie(list(dct.values()), autopct='%0.2f',labels=dct.keys())
    plt.pie(dct.values())
    plt.xlabel("Month")
    plt.ylabel("Jobs")
    plt.title('Month vs Job_Openings')
    plt.xticks(rotation = 40)
    plt.show()
```
