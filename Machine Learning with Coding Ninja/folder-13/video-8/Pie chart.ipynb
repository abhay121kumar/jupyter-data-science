```python
import matplotlib.pyplot as plt
```

```python
sizes = [3,4,6,2]
plt.pie(sizes)
plt.axis("equal")
```

```python
sizes = [3,4,6,2]
colors = ["black", "gray", "pink", "yellow"]
plt.pie(sizes, colors=colors)
plt.axis("equal")
```

```python
sizes = [3,4,6,2]
labels = ["a", "b", "c", "d"]
colors = ["black", "gray", "pink", "yellow"]
plt.pie(sizes, colors=colors, labels=labels)
plt.axis("equal")
```

```python
sizes = [3,4,6,2]
labels = ["a", "b", "c", "d"]
plt.title("split among classes")
colors = ["black", "gray", "pink", "yellow"]
plt.pie(sizes, colors=colors, labels=labels)
plt.axis("equal")
```

```python
sizes = [3,4,6,2]
labels = ["a", "b", "c", "d"]
plt.title("split among classes")
colors = ["brown", "gray", "pink", "yellow"]

plt.pie(sizes, colors=colors, labels=labels, autopct="%.1f")

plt.axis("equal")
```

```python
sizes = [3,4,6,2]
labels = ["a", "b", "c", "d"]
plt.title("split among classes")
colors = ["black", "gray", "pink", "yellow"]
plt.pie(sizes, colors=colors, labels=labels, autopct="%.3f")
plt.axis("equal")
```

```python
sizes = [3,4,6,2]
labels = ["a", "b", "c", "d"]
plt.title("split among classes")
colors = ["black", "gray", "pink", "yellow"]
plt.pie(sizes, colors=colors, labels=labels, autopct="%.2f%%")
plt.axis("equal")
```

```python
sizes = [3,4,6,2]
labels = ["a", "b", "c", "d"]
plt.title("split among classes")
colors = ["black", "gray", "pink", "yellow"]

explode = [0.1,0,0,0]

plt.pie(sizes, colors=colors,explode=explode, labels=labels, autopct="%.2f%%")
plt.axis("equal")
```

```python
sizes = [3,4,6,2]
labels = ["a", "b", "c", "d"]
plt.title("split among classes")
colors = ["black", "gray", "pink", "yellow"]

explode = [0.1,0.2,0,0]

plt.pie(sizes, colors=colors,explode=explode, labels=labels, autopct="%.2f%%")
plt.axis("equal")
```

```python
sizes = [3,4,6,2]
labels = ["a", "b", "c", "d"]
plt.title("split among classes")
colors = ["black", "gray", "pink", "yellow"]

explode = [0.1,0.2,0,0]

plt.pie(sizes, colors=colors,explode=explode, labels=labels, autopct="%.2f%%", counterclock=False)
plt.axis("equal")
```

```python
sizes = [3,4,6,2]
labels = ["a", "b", "c", "d"]
plt.title("split among classes")
colors = ["black", "gray", "pink", "yellow"]

explode = [0.1,0.2,0,0]

plt.pie(sizes, colors=colors,explode=explode, labels=labels, autopct="%.2f%%", counterclock=False, startangle=100)
plt.axis("equal")
```

```python
sizes = [3,4,6,2]
labels = ["a", "b", "c", "d"]
plt.title("split among classes")
colors = ["black", "gray", "pink", "yellow"]

explode = [0.1,0.2,0,0]

plt.pie(sizes, colors=colors,explode=explode, labels=labels, autopct="%.2f%%", startangle=100)
plt.axis("equal")
```
