```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4]
# y=np.array(x**2)
y=[]
for i in range(len(x)):
    y.append(i**2)
plt.plot(x,y)
```

```python
x = np.arange(0,5,0.1)
# y=np.array(x**2)
y=[]
for i in range(len(x)):
    y.append(i**2)
plt.plot(x,y,c="r",marker="o")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("x^2 graph")
```

```python
plt.scatter(x,y,c="r",marker="o")
```
