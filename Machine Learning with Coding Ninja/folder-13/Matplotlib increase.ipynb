```python
import numpy as np
x = np.arange(0,40,2)
y = []
for i in range(len(x)):
    y.append(2**x[i])
y = np.array(y)
print(x)
print(y)
plt.plot(x, y, "b--")
print()
print(30,"to", 35)
print("x", 30)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,40,2)
y = 2**x

plt.plot(x,y,'--',color ='blue')
plt.grid()
plt.show()
print(30)
```
