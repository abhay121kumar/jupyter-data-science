```python
import numpy as np
import matplotlib.pyplot as plt
company = ['HP','Dell','Lenovo','Asus','Apple','Acer']
sold = [20000,43000,15000,17000,22000,13000]
```

```python
plt.title("laptop sales")
plt.pie(sold,labels=company,autopct="%.1f")

company = np.array(company)
sold = np.array(sold)

x = np.true_divide(sold, sold.sum(axis=0, keepdims=True))*100

for i in range(len(company)):
    print(company[i],"->" ,format(x[i], '.1f'))
```

```python
import numpy as np
import matplotlib.pyplot as plt
company = ['HP','Dell','Lenovo','Asus','Apple','Acer']
sold = [20000,43000,15000,17000,22000,13000]

company = np.array(company)
sold = np.array(sold)

plt.pie(sold, labels=company, autopct='%1.1f%%', counterclock=False)
plt.axis('equal')
plt.show()

x = np.true_divide(sold, sold.sum(axis=0, keepdims=True))*100
for i in range(len(company)):
    print(company[i],format(x[i],'.1f'))
```

```python
import matplotlib.pyplot as plt

def fun(sold, company):
    li = []
    sum1 = sum(sold)
    for i in range(len(sold)):
        num = (sold[i]*100)/sum1
        print(company[i], round(num, 1))
        li.append(num)
    s = str(li)
    return s
plt.axis("equal")
company = ['HP','Dell','Lenovo','Asus','Apple','Acer']
sold = [20000,43000,15000,17000,22000,13000]
plt.pie(sold, labels=company, explode= [0.1,0.2,0.1,0.2,0.1,0], autopct=fun(sold, company), counterclock=False, startangle=90)
```
