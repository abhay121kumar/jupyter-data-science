```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(year, salary)
```

```python
plt.scatter(year, salary, s=100)
```

```python
plt.scatter(year, salary, s=population)
```

```python
plt.scatter(year, salary, s=population, c='r')
```

```python
plt.title("Bubble chart")
plt.scatter(year, salary, s=population, c='r')
```
