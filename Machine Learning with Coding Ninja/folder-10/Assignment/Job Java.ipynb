```python
import csv

with open('amazon_jobs_dataset.csv') as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace=True)
    file_list = list(file_data)
    
    count=0
    country_dict={}
    for row in file_list:
        key1 = row['BASIC QUALIFICATIONS']
        key2 = row['location']
        file_str = "Java"
        ans = key1.find(file_str)
        if ans != -1:
            if key2 in country_dict.keys():
                country_dict[key2]+=1
            else:
                country_dict[key2] = 1
        else:
            continue
            
#     print( country_dict)
country_dict
```
